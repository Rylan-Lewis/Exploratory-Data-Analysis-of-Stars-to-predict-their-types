```python
plt.boxplot(df['Radius(R/Ro)'])
plt.xlabel('Radius')
plt.ylabel('Range in metres')
plt.title('Boxplot of Relative Radius')
plt.show()
```

Here, there are quite a lot of values indicating a high Relative Radius that fall in the outlier category. There is a reason for this as well. The Relative Radius of a star depends on its Temperature and Relative Luminosity as R ~ sqrrt(L)/T^(2). Now we can see that most of the outliers data here has extemely high Relative Luminosity and low temperatues in the 3000-4000 F range. Thus in these cases, the Relative Radius of the stars are large, being the reason that it can give off more energy and shine brighter as it takes up more space and its range of light energy burst is increased.

##C) Spectral Class and Relative Radius Comparison

```python
x = df['Spectral Class']
y = df['Radius(R/Ro)']
norm = mcolors.Normalize(vmin=min(y), vmax=max(y))
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=y, cmap="YlOrRd", norm=norm, alpha=0.8)
plt.xlabel('Spectral Class')
plt.ylabel('Radius (R/Ro)')
plt.title('Spectral Class vs. Radius')
plt.colorbar(scatter, label='Radius (R/Ro)')
plt.tight_layout()
plt.show()
```

As we can see, The Higher spectral class stars like O, A and B have most stars that have Relative Radiuses in the higher range of values. The reasons for this are these are SuperGiants and Hypergiant stars that  have very very high temperatures and gases like hydrogen and helium on the outer layer of the star exapnd due to the high temperatures thus increaseing the overall Relative Radius. Now, M stars which are Mian sequence stars are usually stars with radii smaller than the sun, but in this scatterplot it is clearly visible that there are many stars in M class having large radii. The reason is concurred below.

```python
df.loc[(df['Spectral Class']=="M") & (df['Radius(R/Ro)']>750) ]
```

```python
df.loc[(df['Spectral Class']=="M") & (df['Absolute magnitude(Mv)']<0)]
```

From the above two operations, we see a common trend:

A) The Star types of these M spectral stars are all 4 or 5, indicating they're all Supergiants and Hypergiants. A supergiant or Hypergiant star can be classified into any of the Spectral Classes O,A,B,M. However, the temperature of all these Stars are well inside the range of M type stars. Thus the only parameter that can confirm this is Relative Luminosity

B) Another pattern that we can notice is for all these M type stars, The Absolute Magnitude is a high negative value. A highly negative Absolute magnitude value indicates that the Star is actually a Very Bright and Luminous star. This could be because of the Chemical composition of the star and the rate at which it releases energy. And we already know Relative Luminosity and Relative Radius is directly proportional, thus facilitating this conculision more.

Therefore, The M type star having High Relative Radius owe it to their Low temperature values(Within the range of M stars) and High negative Absolute Magniute indicating very high Relative Luminosity, which inturn is the reason for its large Relative Radius.

#F) Temperature VS Relative Radius

```python
x= df['Radius(R/Ro)']
y= df['Temperature (K)']
plt.scatter(x,y)
plt.title("Radius VS Temperature")
plt.ylabel("Temperature")
plt.xlabel("Radius")
plt.tight_layout()
plt.show()
```

Radius and Temperature have kind of an inverse relationship as Radius being inversely proportional to the square of temperature. Thus, as it should, as the radius increases, temperature should decrease. We can see that the high temperature stars are more concentrated in the low radii region. These can include Main sequence stars as well as White dwarfs( These are stars that have exhausted their fuel so theyre just hot with a compact radii). The  high temperature stars with high radiuses are actually Supergiants or Hypergiants as they constantly burn their hydrogen and helium fuel, making their temperature rise and pushing energy from the core at a high rate resulting in an increase of the outer layer of the star.

```python
df['Star color'].unique()
```

```python
x=df['Star color'].unique()
print(x)
```

```python
color_temperature = df.groupby('Star color')['Temperature (K)'].mean()
unique_colors = color_temperature.index
mean_temperatures = color_temperature.values
plt.figure(figsize=(10, 6))
plt.bar(unique_colors, mean_temperatures, color="Red")
plt.xlabel('Star Color')
plt.ylabel('Mean Temperature (K)')
plt.title('Mean Temperature of Stars by Star Color')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

As we can see, it is clearly visible that the Blue Color Stars are the stars with the highest mean temperatures above 20000K. These are usally Supergiants and Hypergiants belonging to O, A, B or M spectral classes. The Red and orange color stars are the stars with the lowest temperatures less than 5000K and are usually the main sequence stars belonging to M,K,G spectral classes.
